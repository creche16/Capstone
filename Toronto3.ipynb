First we get the table info like a dataframe

```python
url='http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = requests.get(url).text
df_list = pd.read_html(html)
df=df_list[-0]
toronto=pd.DataFrame(df)
toronto.head()
```

Then we ignore cells with a value of Not assigned in -borough

```python
toronto=toronto[toronto.Borough != 'Not assigned']
toronto.head()
```

Next, we see which rows have Not assigned value on Neighbouhood, the value of the Borough will be equal to the Neighbourhood

```python
for n in range(1,len(toronto)):
    if toronto.iat[n-1,1]=='Not assigned':
        print(toronto(n-1,2))
        toronto.iat[n-1,2]=toronto.iat[n-1,1]
        

    
    
```

Finally, we join all the neighbourhoods with same Postcode and also Borough

```python
toronto=toronto.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
toronto
```

To insert the Latitude and Longitude, qe create a new dataframe called neighbourhoods and we append the information using the toronto df and the information of the postal codes

```python
column_names = ['Postal Code','Borough', 'Neighborhood', 'Latitude', 'Longitude'] 

# instantiate the dataframe
neighborhoods = pd.DataFrame(columns=column_names)
postal_code=pd.read_csv('http://cocl.us/Geospatial_data')
postal_code.head()
```

```python
for n in range(1,len(toronto)):
    pc=toronto.iat[n-1,0]
    borough=toronto.iat[n-1,1]
    nh=toronto.iat[n-1,2]
    toronto_pc=postal_code.loc[postal_code['Postal Code'] == toronto.iat[n-1,0]]
    lat=toronto_pc.iat[0,1]
    lon=toronto_pc.iat[0,2]
    neighborhoods = neighborhoods.append({'Postal Code':pc,
                                          'Borough': borough,
                                          'Neighborhood': nh,
                                          'Latitude': lat,
                                          'Longitude': lon}, ignore_index=True)
neighborhoods.head()
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="t_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
